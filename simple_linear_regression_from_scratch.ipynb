```python
import numpy as np 

class LinearRegressionFromScratch:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.learning_rate = learning_rate  # step size for gradient updates
        self.n_iters = n_iters              # number of training iterations
        self.w = None                      # slope 
        self.b = None                      # intercept 

    def fit(self,x,y):
        X = np.array(x).reshape(-1)         # convert input to 1D numpy array
        y = np.array(y).reshape(-1)         # convert target to 1D numpy array
        m = len(X)                          # number of training samples
        
        self.w = 0.0                       # initialize slope to 0
        self.b = 0.0                       # initialize intercept to 0

        #Gradient Descent loop 
        for _ in range(self.n_iters):
            y_pred = self.b + self.w*X    # compute predictions
            error = y_pred-y                # compute error (pred - actual)

            grad_w = (1/m)* np.dot(error,X) # gradient of cost w.r.t w
            grad_b = (1/m)* np.sum(error)   # gradient of cost w.r.t b

            self.w -= self.learning_rate*grad_w  # update slope
            self.b -= self.learning_rate*grad_b  # update intercept


    def predict(self,X):
        X = np.array(X).reshape(-1)         # convert input to 1D numpy array
        return self.w*X +self.b           # apply y = w*X + b
```

```python
if __name__ == "__main__":
    # this block only runs when the file is executed directly,
    # not when it's imported as a module elsewhere

    # quick sanity test
    X = [1, 2, 3, 4, 5]           # sample input feature values
    y = [2, 4, 5, 4, 5]           # sample target values

    model = LinearRegressionFromScratch(learning_rate=0.01, n_iters=1000)  # create model instance
    model.fit(X, y)               # train the model on sample data

    print("Learned slope (w):", model.w)              # print final learned slope
    print("Learned intercept (b):", model.b)           # print final learned intercept
    print("Prediction for X=6:", model.predict([6]))   # test prediction on new input
```
